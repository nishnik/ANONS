```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torchvision import datasets,transforms
import matplotlib.pyplot as plt
import pandas as pd
```

```python
transform = transforms.Compose([transforms.ToTensor(),
                                 transforms.Normalize((0.76487684, 0.75205952, 0.74630833), (0.27936298, 0.27850413, 0.28387958))])
```

```python
trainset = datasets.CIFAR10(root='./data', train=True,
                                        download=True, transform=transform)
train_loader = torch.utils.data.DataLoader(trainset, batch_size=16,
                                          shuffle=True)

testset = datasets.CIFAR10(root='./data', train=False,
                                       download=True, transform=transform)
test_loader = torch.utils.data.DataLoader(testset, batch_size=16)
```

```python
classes = ('plane', 'car', 'bird', 'cat',
           'deer', 'dog', 'frog', 'horse', 'ship', 'truck')
```

```python
class Classifier(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1=nn.Linear(3072,1536)
        self.fc2=nn.Linear(1536,384)
        self.fc3=nn.Linear(384,192)
        self.fc4=nn.Linear(3072,768)
        self.fc5=nn.Linear(768,192)
        self.fc6=nn.Linear(192,10)
        # Dropout module with 0.2 drop probability
        self.dropout=nn.Dropout(p=0.2)
    def forward(self,x,a):
        # make sure input tensor is flattened
        x = x.view(x.shape[0], -1)

        # Now with dropout
        if(a==4):
            x=self.dropout(F.relu(self.fc1(x)))
            x=self.dropout(F.relu(self.fc2(x)))
            x=self.dropout(F.relu(self.fc3(x)))
        if(a==3):
            x=self.dropout(F.relu(self.fc4(x)))
            x=self.dropout(F.relu(self.fc5(x)))
        # output so no dropout here
        x = F.log_softmax(self.fc6(x), dim=1)

        return x
```

```python
data=pd.DataFrame(columns=['Learning_Rate','Momentum','No_of_layer','epoch1','epoch2','epoch3','epoch4','epoch5','epoch6','epoch7','epoch8','epoch9','epoch10','Mean_accuracy'],index=range(1,25))
```

```python
model=Classifier()
lr=[0.001,0.004,0.008,0.01]
momentum=[0.3,0.6,0.9]
overall_acu=[]
d=-1
for l in lr:
  for m in momentum:
    for hl in range(3,5):
        d+=1
        data.iloc[d][0]=l
        data.iloc[d][1]=m
        data.iloc[d][2]=hl
        er=2
        print("  -----------------For learning_Rate:{}  Momentum:{}  No_of_Layer:{}----------------".format(l,m,hl))
        criterian=nn.NLLLoss()
        optimizer=optim.SGD(model.parameters(),lr=l,momentum=m)
        acuracy=[]
        epoch=10
        train_losses=[]
        test_losses=[]
        for e in range(epoch):
            running_loss=0
            er+=1
            for images,labels in train_loader:
                images=images.view(images.shape[0],-1)
                optimizer.zero_grad()
                output=model(images,hl)
                loss=criterian(output,labels)
                loss.backward()
                optimizer.step()
                running_loss+=loss
            else:
                test_loss=0
                aaccuracy=0
                with torch.no_grad():
                    model.eval()
                    for images,labels in test_loader:
                        images=images.view(images.shape[0],-1)
                        log_ps=model(images,hl)
                        test_loss+=criterian(log_ps,labels)
                        ps=torch.exp(log_ps)
                        top_p,top_class=ps.topk(1,dim=1)
                        equals=top_class==labels.view(*top_class.shape)
                        aaccuracy+=torch.mean(equals.type(torch.FloatTensor))
                    aaccuracy=aaccuracy/len(test_loader)
                    acuracy.append(aaccuracy)
                model.train()
                data.append
                print("Epoch: {}/{}.. ".format(e+1, epoch),
                      "Training Loss: {:.3f}.. ".format(running_loss/len(train_loader)),
                      "Test Loss: {:.3f}.. ".format(test_loss/len(test_loader)),
                      "Test Accuracy: {:.3f}".format(aaccuracy))
                data.iloc[d][er]=aaccuracy
            train_losses.append(running_loss/len(train_loader))
            test_losses.append(test_loss/len(test_loader))
        overall_acu.append(sum(acuracy)/epoch)
        data.iloc[d][er+1]=sum(acuracy)/epoch
        print("accuracy mean: {:.3f}".format(sum(acuracy)/epoch))
        fig=plt.figure(figsize=(11,6))
        plt.plot(train_losses, label='Training loss')
        plt.plot(test_losses, label='Validation loss')
        plt.ylabel('Loss')
        plt.xlabel('Epoch-1')
        plt.title("Learning_Rate:{},Momentum:{},No_Layer:{},Mean_ACcuracy:{:.3f}".format(l,m,hl,sum(acuracy)/epoch))    
        plt.legend()
        plt.show()
        fig.savefig("LearningRate{} Momentum{} NO_LAyer{}.png".format(l,m,hl))
```

```python
data.to_csv('data.csv')
```

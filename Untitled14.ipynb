```python
import torch
from torch import nn
from torch import optim
import torch.nn.functional as F
from torchvision import datasets,transforms
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import matplotlib.pyplot as plt
```

```python
transform=transforms.Compose([transforms.ToTensor(),
                            transforms.Normalize((0.5,0.5,0.5),(0.5,0.5,0.5))])
```

```python
trainset=datasets.MNIST('~/.pytorch/MNIST_data/',download=True,train=True,transform=transform)
trainloder=torch.utils.data.DataLoader(trainset,batch_size=64,shuffle=True)

testset=datasets.MNIST('~/.pytorch/MNIST_data/',download=True,train=False,transform=transform)
testloder=torch.utils.data.DataLoader(testset,batch_size=64,shuffle=True)
```

```python
class Classifier(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(784, 256)
        self.fc2 = nn.Linear(256, 64)
        self.fc3 = nn.Linear(64, 10)

        # Dropout module with 0.2 drop probability
        self.dropout = nn.Dropout(p=0.2)

    def forward(self, x,a,b):
        # make sure input tensor is flattened
        x = x.view(x.shape[0], -1)

        # Now with dropout
        x = self.dropout(a(self.fc1(x)))
        x = self.dropout(b(self.fc2(x)))

        # output so no dropout here
        x = F.log_softmax(self.fc3(x), dim=1)

        return x
```

```python
f=0
model=Classifier()
overall_acu=[]
lrr=[0.03,0.003,0.001]
for x in lrr:
    print(f"                         ----------------FOR LEARNING RATE {x}--------------------                        ")
    for a in[nn.ReLU(),nn.Tanh(),nn.Sigmoid(),nn.SELU()]:
        for b in[nn.ReLU(),nn.Tanh(),nn.Sigmoid(),nn.SELU()]:
            print("Layer1--{}     Layer2--{}".format(a,b))
            f+=1
            criterian=nn.NLLLoss()
            optimizer=optim.SGD(model.parameters(),lr=x)
            aaccuracy=0
            epoch=7
            train_losses=[]
            test_losses=[]
            for e in range(epoch):
                running_loss=0
                for images,labels in trainloder:
                    images=images.view(images.shape[0],-1)
                    optimizer.zero_grad()
                    output=model(images,a,b)
                    loss=criterian(output,labels)
                    loss.backward()
                    optimizer.step()
                    running_loss+=loss.item()
                else:
                    test_loss=0
                    accuracy=0
                    with torch.no_grad():
                        model.eval()
                        for images,labels in testloder:
                            images=images.view(images.shape[0],-1)
                            log_ps=model(images,a,b)
                            test_loss+=criterian(log_ps,labels)
                            ps=torch.exp(log_ps)
                            top_p,top_class=ps.topk(1,dim=1)
                            equals=top_class==labels.view(*top_class.shape)
                            accuracy+=torch.mean(equals.type(torch.FloatTensor))
                        accuracy=(accuracy/len(testloder))
                        aaccuracy+=accuracy
                    model.train()
                    print("Epoch: {}/{}.. ".format(e+1, epoch),
                          "Training Loss: {:.3f}.. ".format(running_loss/len(trainloder)),
                          "Test Loss: {:.3f}.. ".format(test_loss/len(testloder)),
                          "Test Accuracy: {:.3f}".format(accuracy))
                train_losses.append(running_loss/len(trainloder))
                test_losses.append(test_loss/len(testloder))
            aaccuracy=aaccuracy/epoch
            overall_acu.append(aaccuracy)
            print("accuracy mean: {:.3f}".format(aaccuracy))
            fig=plt.figure(figsize=(11,6))
            plt.plot(train_losses, label='Training loss')
            plt.plot(test_losses, label='Validation loss')
            plt.ylabel('Loss')
            plt.xlabel('Epoch-1')
            plt.title("Layer1:{},Layer2:{},Mean_ACcuracy:{:.3f}".format(a,b,aaccuracy))      
            plt.legend()
            plt.show()
            fig.savefig("{}.png".format(f))
```

```python
accu=overall_acu
```

```python
for i in range(len(accu)):
    accu[i]=accu[i].numpy()
```

```python
import pandas as pd
```

```python
result=pd.DataFrame(columns=['Learning Rate','Layer1','Layer2','Accuracy_7_epoch'],index=range(1,len(accu)+1))
```

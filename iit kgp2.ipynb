```python
import torch
from torch import nn
from torch import optim
from torchvision import datasets,transforms
import torch.nn.functional as F
import matplotlib.pyplot as plt
import pandas as pd
```

```python
transform=transforms.Compose([transforms.ToTensor(),
                             transforms.Normalize((0.5,0.5,0.5),(0.5,0.5,0.5))])
```

```python
trainset=datasets.MNIST('~/.pytorch/MNIST_data',train=True,transform=transform,download=True)
trainloder=torch.utils.data.DataLoader(trainset,batch_size=64)

testset=datasets.MNIST('~/.pytorch/MNIST_data',train=False,transform=transform,download=True)
testloder=torch.utils.data.DataLoader(testset,batch_size=64)
```

```python
class Classifier(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1=nn.Linear(784,256)
        self.fc2=nn.Linear(784,64)
        self.fc3=nn.Linear(256,64)
        self.fc4=nn.Linear(64,10)
        # Dropout module with 0.2 drop probability
        self.dropout=nn.Dropout(p=0.2)
    def forward(self,x,a):
        # make sure input tensor is flattened
        x = x.view(x.shape[0], -1)

        # Now with dropout
        if(a==3):
            x=self.dropout(F.relu(self.fc1(x)))
            x=self.dropout(F.relu(self.fc3(x)))
        if(a==2):
            x=self.dropout(F.relu(self.fc2(x)))
        # output so no dropout here
        x = F.log_softmax(self.fc4(x), dim=1)

        return x
```

```python
data=pd.DataFrame(columns=['Learning_Rate','No_of_layer','epoch1','epoch2','epoch3','epoch4','epoch5','epoch6','epoch7','epoch8','epoch9','epoch10','Mean_accuracy'],index=range(1,11))
```

```python
f=300
model=Classifier()
lr=[0.001,0.01,0.05,0.1,0.5]
overall_acu=[]
d=-1
for l in lr:
    for hl in range(2,4):
        d+=1
        data.iloc[d][0]=l
        data.iloc[d][1]=hl
        er=1
        print("  -----------------For learning_Rate:{}    No_of_Layer:{}----------------".format(l,hl))
        f+=1
        criterian=nn.NLLLoss()
        optimizer=optim.SGD(model.parameters(),lr=l)
        acuracy=[]
        epoch=10
        train_losses=[]
        test_losses=[]
        for e in range(epoch):
            running_loss=0
            er+=1
            for images,labels in trainloder:
                images=images.view(images.shape[0],-1)
                optimizer.zero_grad()
                output=model(images,hl)
                loss=criterian(output,labels)
                loss.backward()
                optimizer.step()
                running_loss+=loss
            else:
                test_loss=0
                aaccuracy=0
                with torch.no_grad():
                    model.eval()
                    for images,labels in testloder:
                        images=images.view(images.shape[0],-1)
                        log_ps=model(images,hl)
                        test_loss+=criterian(log_ps,labels)
                        ps=torch.exp(log_ps)
                        top_p,top_class=ps.topk(1,dim=1)
                        equals=top_class==labels.view(*top_class.shape)
                        aaccuracy+=torch.mean(equals.type(torch.FloatTensor))
                    aaccuracy=aaccuracy/len(testloder)
                    acuracy.append(aaccuracy)
                model.train()
                data.append
                print("Epoch: {}/{}.. ".format(e+1, epoch),
                      "Training Loss: {:.3f}.. ".format(running_loss/len(trainloder)),
                      "Test Loss: {:.3f}.. ".format(test_loss/len(testloder)),
                      "Test Accuracy: {:.3f}".format(aaccuracy))
                data.iloc[d][er]=aaccuracy
            train_losses.append(running_loss/len(trainloder))
            test_losses.append(test_loss/len(testloder))
        overall_acu.append(sum(acuracy)/epoch)
        data.iloc[d][er+1]=sum(acuracy)/epoch
        print("accuracy mean: {:.3f}".format(sum(acuracy)/epoch))
        fig=plt.figure(figsize=(11,6))
        plt.plot(train_losses, label='Training loss')
        plt.plot(test_losses, label='Validation loss')
        plt.ylabel('Loss')
        plt.xlabel('Epoch-1')
        plt.title("Learning_Rate:{},Mean_ACcuracy:{:.3f}".format(l,sum(acuracy)/epoch))    
        plt.legend()
        plt.show()
        fig.savefig("{}.png".format(f))
```
